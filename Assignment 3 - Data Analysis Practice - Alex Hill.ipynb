### Figure 2

```python
x = np.load("Assignment 3 data/data-fig2.npy")
#print(x)

plt.figure(figsize=(4.4,2.8))

plt.plot(x[:,0], color=(1,0,0), linewidth=1)
plt.plot(x[:,1], color="g", linewidth=1)
plt.plot(x[:,2], color=(0,0,1), linewidth=1)


plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.legend(["Red","Green","Blue"])

#plt.title("Title")
#plt.xticks([])
#ax= plt.gca()
#ax.set_xticklabels([])

plt.show()


```

### Figure 4

```python
plt.figure(figsize=(4.4,2.8))

y_pos= ('A', 'B', 'C', 'D', 'E')
x = np.load("Assignment 3 data/data-fig4.npy")

plt.bar(y_pos, x, align='center', alpha=0.8,color=(0.6,0.6,0.6))
plt.xticks(y_pos)
plt.ylabel('Value')
plt.xlabel('Category')


plt.show()
```
